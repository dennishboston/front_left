```python
flightId = []
dates = []
delays = []
irregularOperations = []

for status in response["flightStatuses"]:
    #print(status["delays"]["departureRunwayDelayMinutes"])
    dates.append(status["departureDate"]["dateUtc"])
    flightId.append(status["flightId"])
    #delays.append(status["delays"]["departureRunwayDelayMinutes"])
    try:
        if status["irregularOperations"][0]["type"] != "":
            #print(status["irregularOperations"][0]["type"])
            irregularOperations.append(status["irregularOperations"][0]["type"])
    except Exception:
        irregularOperations.append("None")
        

        
#make df from dicts, label as ord
ord_data = pd.DataFrame({"Flight Id": flightId, "Dates" : dates, "Irregular Ops" :irregularOperations})


ord_data.head()
```
